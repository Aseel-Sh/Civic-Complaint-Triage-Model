# 02 - Exploratory Analysis

We explore the cleaned dataset and save key charts to `reports/figures/`.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd

project_root = Path('..').resolve()
data_path = project_root / 'data' / 'processed' / 'complaints_cleaned.csv'
figures_dir = project_root / 'reports' / 'figures'
figures_dir.mkdir(parents=True, exist_ok=True)

df = pd.read_csv(data_path)
df.head()
```

## Complaint volume by month
This shows seasonality in complaint volume.

```python
if 'submitted_month' in df.columns:
    monthly = df['submitted_month'].value_counts().sort_index()
    plt.figure(figsize=(7, 4))
    monthly.plot(kind='bar', color='#4C78A8')
    plt.title('Complaint Volume by Month')
    plt.xlabel('Month')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.savefig(figures_dir / 'complaints_by_month.png', dpi=150)
    plt.show()
```

## Top complaint types
We only plot this if `complaint_type` is available.

```python
if 'complaint_type' in df.columns:
    top_types = df['complaint_type'].value_counts().head(10)
    plt.figure(figsize=(8, 4))
    top_types.sort_values().plot(kind='barh', color='#F58518')
    plt.title('Top 10 Complaint Types')
    plt.xlabel('Count')
    plt.ylabel('Type')
    plt.tight_layout()
    plt.savefig(figures_dir / 'top_complaint_types.png', dpi=150)
    plt.show()
```

## Delayed vs not delayed
This shows the class balance.

```python
if 'delayed' in df.columns:
    counts = df['delayed'].value_counts().sort_index()
    plt.figure(figsize=(5, 4))
    counts.plot(kind='bar', color=['#54A24B', '#E45756'])
    plt.title('Delayed vs Not Delayed')
    plt.xlabel('Delayed')
    plt.ylabel('Count')
    plt.xticks([0, 1], ['Not delayed', 'Delayed'], rotation=0)
    plt.tight_layout()
    plt.savefig(figures_dir / 'delayed_counts.png', dpi=150)
    plt.show()
```

## Distribution of resolution time
We visualize how long it takes to resolve complaints.

```python
if 'days_to_resolution' in df.columns:
    plt.figure(figsize=(7, 4))
    df['days_to_resolution'].plot(kind='hist', bins=40, color='#72B7B2')
    plt.title('Distribution of Days to Resolution')
    plt.xlabel('Days')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.savefig(figures_dir / 'days_to_resolution_distribution.png', dpi=150)
    plt.show()
```

## Median resolution time by complaint type
This is useful to see which complaint types take longer to resolve.

```python
if 'complaint_type' in df.columns and 'days_to_resolution' in df.columns:
    medians = df.groupby('complaint_type')['days_to_resolution'].median().sort_values(ascending=False).head(10)
    plt.figure(figsize=(8, 4))
    medians.sort_values().plot(kind='barh', color='#B279A2')
    plt.title('Top 10 Complaint Types by Median Resolution Time')
    plt.xlabel('Median Days')
    plt.ylabel('Type')
    plt.tight_layout()
    plt.savefig(figures_dir / 'median_resolution_by_type.png', dpi=150)
    plt.show()
```

## Complaint count by ZIP code
We only plot this if a ZIP code column is available.

```python
if 'zip_code' in df.columns:
    top_zips = df['zip_code'].value_counts().head(10)
    plt.figure(figsize=(8, 4))
    top_zips.sort_values().plot(kind='barh', color='#FF9DA6')
    plt.title('Top 10 ZIP Codes by Complaint Volume')
    plt.xlabel('Count')
    plt.ylabel('ZIP Code')
    plt.tight_layout()
    plt.savefig(figures_dir / 'complaints_by_zip.png', dpi=150)
    plt.show()
```
